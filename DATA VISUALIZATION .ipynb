```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cities=pd.read_csv('california_cities.csv')
```

```python
cities.head()
```

```python
# Extract information
lat,lon=cities['latd'],cities['longd']
```

```python
lat[:10]
```

```python
pop,area=cities['population_total'],cities['area_total_km2']
plt.style.use('seaborn')

plt.figure(figsize=(8,6))
plt.scatter(lon,lat,c=np.log10(pop),cmap='viridis',s=area,linewidths=0,alpha=0.5)

plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(pop)')
plt.clim(3,7);
plt.title('California_cities : Population and area Distribution')
#create legend for city side
area_range=[50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area, label=str(area)+'km$^2$',c='k',alpha=0.4)
plt.legend(labelspacing=1,scatterpoints=1,title="City Area")
```
