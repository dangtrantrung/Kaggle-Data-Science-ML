```python
# import libraries
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv',index_col='Id')
```

```python
data
```

```python
data.columns
```

## Feature Selection, use "Feature Engineering"to select features needed for train Model

```python
features=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
```

# Spliting Dataset into X and Y

```python
X= data[features]
```

```python
X
```

```python
Y=data['SalePrice']
```

```python
Y
```

```python
# X_train, Y_train, X_Valid, Y_Valid
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_Valid,y_train,y_Valid=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
# 4. Training ML model
```

```python
from sklearn.tree import DecisionTreeRegressor 
```

```python
dt_model=DecisionTreeRegressor(random_state=1)
```

```python
# Fit traing data into Model
```

```python
dt_model.fit(X_train,y_train)
```

```python
y_preds=dt_model.predict(X_Valid.head())
```

```python
pd.DataFrame({'y':y_Valid.head(),'y_preds':y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor 
```

```python
rf_model=RandomForestRegressor(random_state=1)
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_val_preds=rf_model.predict(X_Valid)
```

```python
pd.DataFrame({'y':y_Valid,'y_preds':rf_val_preds})
```
